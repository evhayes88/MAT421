## 9.1 Base-N and Binary

### Base-N

Generically, a number can be represented by a base-N system.

Given a base-N system,

*   A number is represented by a list of integer digits from $0$ to $N-1$.

*   Each integer digit represents a coefficient for a power of $N$.

*   A number of base-N is denoted with a subscript $N$.

**Example 1.** Show the base-N expansion for $123_N$.

$ 123_N = 1⋅N^2 + 2⋅N^1 + 3⋅N^0 $

The most common system is the decimal system, also known as base-10.

**Example 2.** Show the base-10 expansion for $101.1_{10}$.

$ 101.1_{10} = 1⋅10^2 + 0⋅10^1 + 1⋅10^0 + 1⋅10^{-1} $

### Binary

Computers use the binary system, or base-2 system.

**Example 3.** Show the base-2 expansion for $7_{10}$.

$ 7_{10} = 1⋅2^2 + 1⋅2^1 + 1⋅2^0 = 111 $








## 9.2 Floating Point Numbers

Binary representation of numbers have a downside. They can only represent positive integers. As a result, a standard was created in order to represent fractional numbers in a computer. The standard is IEEE754. A fractional number will be represented by the following parts:

*   Sign Indicator
*   Exponent
*   Fraction

Depending on the precision of the number, a certain amount of bits will be allocated to each of the listed parts.





## 9.3 Round-off Errors

### Representation Error

Representation error is due to the fact that a computer has a finite amount of bits and numbers such as $\pi$ or $1/3$ would require an infinite amount. As such, there will be an error between the floating point representation and it's theoritical value. 

### Error by Floating-Point Arithmetic

Numbers are often approximated as a floating point number and will then contain a very small error. When arithmetic is applied these errors can cause issues in a program. It is desirable to round values or have precision thresholds.

### Accumulation of Round-Off Error

When a program iterates through multiple calculations the error from arithmetic can accumulate leading to large errors in a program.
